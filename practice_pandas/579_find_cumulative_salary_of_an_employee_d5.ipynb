```python
import pandas as pd
```

```python
data = [
    {"Id": 1, "Month": 1, "Salary": 20},
    {"Id": 2, "Month": 1, "Salary": 20},
    {"Id": 1, "Month": 2, "Salary": 30},
    {"Id": 2, "Month": 2, "Salary": 30},
    {"Id": 3, "Month": 2, "Salary": 40},
    {"Id": 1, "Month": 3, "Salary": 40},
    {"Id": 3, "Month": 3, "Salary": 60},
    {"Id": 1, "Month": 4, "Salary": 60},
    {"Id": 3, "Month": 4, "Salary": 70},
]
```

```python
employee_df=pd.DataFrame(data)
employee_df
```

```python
employee_df=employee_df.sort_values(['Id','Month'])

employee_df['running_total']=employee_df.groupby('Id')['Salary'].cumsum()
employee_df
```

```python
employee_df=employee_df.sort_values('running_total',ascending=False)
employee_df['row_num']=employee_df.groupby('Id')['running_total'].cumcount() + 1
employee_df
```

```python
employee_df=employee_df[(employee_df['row_num']!=1) & (employee_df['row_num']<=4)]
employee_df=employee_df.sort_values(['Id','Salary'],ascending=[True,False])
employee_df=employee_df[['Id','Month','running_total']]
employee_df=employee_df.rename(columns={'running_total':'salary'})
employee_df
```
