```python
import pandas as pd

data = [
    ("Aron", "F", "2020-01-01", 17),
    ("Alice", "F", "2020-01-07", 23),
    ("Bajrang", "M", "2020-01-07", 7),
    ("Khali", "M", "2019-12-25", 11),
    ("Slaman", "M", "2019-12-30", 13),
    ("Joe", "M", "2019-12-31", 3),
    ("Jose", "M", "2019-12-18", 2),
    ("Priya", "F", "2019-12-31", 23),
    ("Priyanka", "F", "2019-12-30", 17)
]

columns = ["player_name", "gender", "day", "score_points"]
scores_df = pd.DataFrame(data, columns=columns)
scores_df["day"] = pd.to_datetime(scores_df["day"])
```

```python
scores_df=scores_df.sort_values(['gender','day'])
scores_df
```

```python
scores_df['total']=scores_df.groupby('gender')['score_points'].cumsum()
```
