```python
import pandas as pd
```

```python
data = {
    "transaction_id": [8, 9, 1, 5, 6],
    "day": [
        "2021-04-03 15:57:28",
        "2021-04-28 08:47:25",
        "2021-04-29 13:28:30",
        "2021-04-28 16:39:59",
        "2021-04-29 23:39:28"
    ],
    "amount": [57, 21, 58, 40, 58]
}
```

```python
transaction_df=pd.DataFrame(data)
transaction_df['day']=pd.to_datetime(transaction_df['day'])
```

```python
transaction_df['day_1']=transaction_df['day'].dt.date
transaction_df
```

```python
transaction_df=transaction_df.sort_values(['day_1','amount'],ascending=[True,False])
transaction_df
```

```python
transaction_df['rnk']=transaction_df.groupby('day_1')['amount'].rank(method='min',ascending=False)
transaction_df
```

```python
transaction_df[transaction_df['rnk']==1][['transaction_id']].sort_values('transaction_id').reset_index(drop=True)
```
